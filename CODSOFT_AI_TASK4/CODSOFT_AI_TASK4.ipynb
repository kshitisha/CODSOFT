```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
# Convert string columns to lists before concatenating
movies['genres'] = movies['genres'].apply(lambda x: x.split() if isinstance(x, str) else x)
movies['keywords'] = movies['keywords'].apply(lambda x: x.split() if isinstance(x, str) else x)
movies['cast'] = movies['cast'].apply(lambda x: x.split() if isinstance(x, str) else x)

# Now you can concatenate
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```
